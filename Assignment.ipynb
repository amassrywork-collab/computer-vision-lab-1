### Q3: Boolean Indexing
Given the array: `arr = np.array([12, 45, 23, 67, 34, 89, 56, 21])`

1. Create a new array containing only elements between 30 and 70 (inclusive)
2. Replace all elements greater than 50 with the value 50

````python
# Your solution for Q3 here

# Given array
arr = np.array([12, 45, 23, 67, 34, 89, 56, 21])

# Step 1: Select elements between 30 and 70 (inclusive)
between_30_70 = arr[(arr >= 30) & (arr <= 70)]

# Step 2: Replace elements > 50 with 50
modified_arr = arr.copy()
modified_arr[modified_arr > 50] = 50

# Display results
between_30_70, modified_arr
````

### Q8: Grayscale Conversion
Convert the penguin image to grayscale using three different methods:
1. Simple average method: `(R + G + B) / 3`
2. Weighted method: `0.299R + 0.587G + 0.114B`
3. Lightness method: `(max(R,G,B) + min(R,G,B)) / 2`

Display all three results and compare them.

````python
from skimage.color import rgb2gray

# Load the image
penguin = image.imread('penguin.jpg')

# ===============================
# Method 1: Simple Average
# (R + G + B) / 3
# ===============================
gray_avg = (penguin[:, :, 0] +
            penguin[:, :, 1] +
            penguin[:, :, 2]) / 3

# ===============================
# Method 2: Weighted Average (Luminosity)
# 0.299R + 0.587G + 0.114B
# ===============================
gray_weighted = (
    0.299 * penguin[:, :, 0] +
    0.587 * penguin[:, :, 1] +
    0.114 * penguin[:, :, 2]
)

# ===============================
# Method 3: Lightness Method
# (max(R,G,B) + min(R,G,B)) / 2
# ===============================
max_rgb = np.max(penguin[:, :, :3], axis=2)
min_rgb = np.min(penguin[:, :, :3], axis=2)
gray_lightness = (max_rgb + min_rgb) / 2

# ===============================
# Method 4: Built-in (skimage)
# ===============================
gray_skimage = rgb2gray(penguin)

# ===============================
# Display Results
# ===============================
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(gray_avg, cmap='gray')
plt.title("Simple Average")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(gray_weighted, cmap='gray')
plt.title("Weighted (Luminosity)")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(gray_lightness, cmap='gray')
plt.title("Lightness")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(gray_skimage, cmap='gray')
plt.title("rgb2gray (skimage)")
plt.axis("off")

plt.tight_layout()
plt.show()
````

### Q9: Image Transformation
Create a new image by:
1. Taking the left half of the original penguin image
2. Flipping it horizontally
3. Stacking it with the right half to create a symmetric image

````python
# Step 1: Get the left half of the image
h, w = penguin.shape[0], penguin.shape[1]
mid = w // 2
left_half = penguin[:, :mid]

# Step 2: Flip it horizontally
left_flipped = np.fliplr(left_half)

# Step 3: Get the right half of the image
right_half = penguin[:, mid:]

# Step 4: Stack left-flipped and right halves horizontally
symmetric = np.hstack((left_flipped, right_half))

# Display the result
plt.figure(figsize=(10, 5))
plt.imshow(symmetric)
plt.title("Symmetric Penguin")
plt.axis("off")
plt.show()
````
